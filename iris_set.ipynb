```python
from sklearn import datasets
iris = datasets.load_iris()
import matplotlib.pyplot as plt
fig1 = plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
target_names = iris.target_names.tolist()
plt.legend(handles=fig1.legend_elements()[0], labels=target_names)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

#  
X_train, X_test, y_train, y_test = train_test_split(
    iris.data[:, :2], iris.target, test_size=0.3, random_state=42)

#  
model = LinearRegression()
model.fit(X_train, y_train)

#  
y_pred = model.predict(X_test)
```

```python
fig2 = plt.scatter(iris.data[:, 2], iris.data[:, 3], c=iris.target, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Petal length")
plt.ylabel("Petal width")
plt.legend(handles=fig2.legend_elements()[0], labels=target_names)
```
